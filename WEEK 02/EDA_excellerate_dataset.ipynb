Data Quality Issue: Negative Application Delay

During EDA, negative values were found in the days_to_apply feature, where the application date occurred before the learner signup date. This is logically inconsistent and likely caused by data entry errors or system synchronization issues.

```python
mask = df['days_to_apply'] < 0

df['invalid_days_to_apply'] = mask.astype(int)
df.loc[mask, 'days_to_apply'] = pd.NA
```

Temporal Inconsistency Due to System Migration

EDA revealed that a subset of records have application dates earlier than learner signup dates. Further inspection showed that these records share identical opportunity start and end dates, indicating batch-imported opportunities.

This suggests that application timestamps were recorded relative to opportunity lifecycle, while learner signup timestamps reflect later account creation or system migration events.
