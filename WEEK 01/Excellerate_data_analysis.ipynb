### Date related all data's are in object format. so converting it to correct datetime format and also removing the time (as it is not so much relevent)

there were also some problem in %m/%d/%Y and %d/%m/%Y format. that isue was also fixed

```python
date_columns = [
    'Learner SignUp DateTime',
    'Date of Birth',
    'Apply Date',
    'Opportunity Start Date',
    'Opportunity End Date',
    'Entry created at'
]
```

```python
for col in date_columns:
    df1[col] = df1[col].str.split(" ").str[0]
```

```python
df3=df1.copy()
```

```python

for col in date_columns:
    df3[col] = pd.to_datetime(
        df3[col],
        format="mixed", # as %m/%d/%Y and %d/%m/%Y formats are mixed
        dayfirst=True,
        errors="coerce"
    )
```

```python
df3.isna().sum()
```

```python
df5 = df3.copy() # converted all dates correctly
```

```python
# removing some columns from these rows having null values
df5 = df5.dropna(
    subset=['Institution Name', 'Current/Intended Major']
)
```

```python
df5.isna().sum()
```

```python
# 8 types of ‘Status Code’ for 8 type of ‘status description’ (co-related) (as they represents each other, so one of them can be removed)
# as "Status Description" is co-related to "Opportunity Start Date", let's drop status code.
df5 = df5.drop(columns=["Status Code"])
```

```python
df5.drop(columns='Opportunity Id',inplace=True)
```

```python
df6 = df5.copy()
```

```python
df6.info()
```

```python
# converting the non-ASCII values(other languages) to english
!pip install unidecode
from unidecode import unidecode
df6["Institution Name"] = df6["Institution Name"].apply(unidecode)
```

```python
df6["First Name"] = df6["First Name"].apply(unidecode)
```

```python
# some object(data type) columns should be converted to lowercase and then check for duplicates. Also should check ‘Whitespace & hidden character’
object_columns = df6.select_dtypes(include=['object']).columns

# Clean object columns (lowercase + whitespace)
df6[object_columns] = (
    df6[object_columns]
    .astype(str)
    .apply(lambda col: col.str.strip().str.lower())
)

"""
What this fixes:

Male → male
FEMALE → female
USA\t → usa
Bangladesh\n → bangladesh"""
```

```python
df6.duplicated().sum()
```

```python
df6 = df6.drop_duplicates() # removing these 6 duplicates
```

```python
df8 = df6.copy()
```

```python
df8.head()
```

```python
df8["Institution Name"] = (
    df8["Institution Name"]
    .astype(str)
    .str.lower()
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
    .str.replace(r"[^a-z0-9\s]", "", regex=True)
)
```

```python
# fixing some errors in Instiuttion Name
corrections = {
    r"univeristy|universitiy": "university",
    r"instiute": "institute",
    r"techonology|technologie": "technology",
    r"coliege|colege": "college",
    r"&": "and",
    r"\bst\.\b|\bst\b": "saint",
    r"\buniv\b|\buni\b": "university",
    r"\binst\b": "institute",
    r"\bengg\b|\beng\b": "engineering",
    r"\btech\b": "technology",
    r"\bint\b": "international",
}


for wrong, correct in corrections.items():
    df8["Institution Name"] = df8["Institution Name"].str.replace(
        wrong, correct, regex=True
    )
```

```python
places = [
    # Bangladesh
    "dhaka", "chittagong", "rajshahi", "khulna", "sylhet",
    "barisal", "rangpur", "mymensingh", "ashulia", "gazipur",
    "tungipara", "gopalganj",

    # Countries
    "bangladesh", "india", "pakistan", "nigeria", "ghana", "egypt",
    "kenya", "uganda", "zimbabwe",

    # Indian states / regions
    "kerala", "punjab", "karnataka", "maharashtra", "tamil nadu",
    "uttar pradesh", "rajasthan", "gujarat", "bihar", "bengal",
    "kashmir", "delhi", "ekiti state", "benue state", "delta state",
    "nassarawa state", "imo state", "ebonyi state", "taraba state",
    "kwara state", "uttarakhand", "ondo state", "kaduna state",

    # Cities (India)
    "mumbai", "pune", "bangalore", "hyderabad", "chennai", "kolkata",
    "nagpur", "jaipur", "lucknow", "patna", "bhopal", "indore",
    "coimbatore", "jodhpur", "noida", "greater noida", "allahabad",
    "nuzvid", "jalgaonmh", "awka", "ogbomoso", "perambalur",

    # Cities (Africa)
    "mardan", "uturu", "bauchi", "karu", "attock", "abraka",
    "effurun", "kano", "lagos", "accra", "cairo", "nairobi",

    # USA / Other cities
    "chicago",

    # Misc noisy suffixes
    "campus", "online", "new", "state"
]


pattern = r"(?<!of\s)\b(" + "|".join(places) + r")\b$"  # don't remove country name in case: "University of Delhi"


df8["Institution Name"] = df8["Institution Name"].str.replace(
    pattern, "", regex=True
).str.strip()
```

```python
df8["Institution Name"].value_counts()
```

```python
# here i am converting "saint louis" to "saint louis university"
df8["Institution Name"] = df8["Institution Name"].replace(
    r"^saint louis$", "saint louis university", regex=True
)
```

```python
df8["Institution Name"].value_counts()
```

```python
df9 = df8.copy()
```

here for "First Name"
1. If a value contains both letters and numbers → remove numbers
2. If a value contains only numbers → replace with "unknown"

```python
# Step 1: replace pure numbers with "unknown"
df9["First Name"] = df9["First Name"].replace(
    r"^\d+$", "unknown", regex=True
)

# Step 2: remove numbers from mixed values
df9["First Name"] = df9["First Name"].str.replace(
    r"\d+", "", regex=True
).str.strip()
```

```python
# marging don't want to specify and other to Unknown
df9["Gender"] = df9["Gender"].replace({
    "don't want to specify": "unknown",
    "other": "unknown"
})
```
